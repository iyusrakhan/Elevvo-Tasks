# Task 4 — Survey Cleaning & Insights

```python

import pandas as pd

# Load the Kaggle Data Science Survey dataset
df = pd.read_csv('/content/kaggle_survey_2017_2021.csv')

# Display shape and first rows
print(df.shape)
df.head()
```

```python

# Check missing values and duplicates
print(df.isnull().sum().head(20))
print("Duplicates:", df.duplicated().sum())
```

```python

# Drop duplicates
df = df.drop_duplicates()

# Handle missing values (example: fill with 'Unknown' for categorical)
df = df.fillna('Unknown')
```

```python

# Encode categorical columns (example: Age and Gender)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
if 'Q1' in df.columns:
    df['Q1_encoded'] = le.fit_transform(df['Q1'].astype(str))
if 'Q2' in df.columns:
    df['Q2_encoded'] = le.fit_transform(df['Q2'].astype(str))

df.head()
```

```python

# Insights Example: Count respondents by gender and country
if 'Q2' in df.columns and 'Q3' in df.columns:
    gender_counts = df['Q2'].value_counts()
    print("Gender Distribution:\n", gender_counts)

    country_counts = df['Q3'].value_counts().head(10)
    print("Top 10 Countries:\n", country_counts)

# Insight: Age distribution
if 'Q1' in df.columns:
    age_counts = df['Q1'].value_counts().sort_index()
    print("Age Distribution:\n", age_counts)

# Insight: Education levels
if 'Q4' in df.columns:
    edu_counts = df['Q4'].value_counts().head(10)
    print("\nTop 10 Education Levels:\n", edu_counts)

# Insight: Employment status
if 'Q5' in df.columns:
    emp_counts = df['Q5'].value_counts().head(10)
    print("\nEmployment Status:\n", emp_counts)


```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(3,2, figsize=(12,10))

# Gender
if 'Q2' in df.columns:
    df['Q2'].value_counts().plot(kind='bar', ax=axs[0,0], title='Gender')

# Age
if 'Q1' in df.columns:
    df['Q1'].value_counts().sort_index().plot(kind='bar', ax=axs[0,1], title='Age')

# Residence
if 'Q3' in df.columns:
    df['Q3'].value_counts().head(5).plot(kind='bar', ax=axs[1,0], title='Top 5 Countries')

# Education
if 'Q4' in df.columns:
    df['Q4'].value_counts().head(5).plot(kind='bar', ax=axs[1,1], title='Education')

# Employment
if 'Q5' in df.columns:
    df['Q5'].value_counts().head(5).plot(kind='bar', ax=axs[2,0], title='Employment')

plt.tight_layout()
plt.show()
```
